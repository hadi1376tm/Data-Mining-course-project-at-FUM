```python
import pandas as pd
```

part 1

```python
raw_data = pd.read_csv("output(2).csv" )
raw_data.drop(["Unnamed: 0"], axis = 1 , inplace = True)
raw_data.tail()
```

```python
raw_data.drop(["Professor_Homepage","Course_Resources"], axis = 1 , inplace = True)
raw_data.tail()
```

```python
raw_data["Departments"][raw_data["Departments"]==" "] = None
raw_data["Course_title"][raw_data["Course_title"]==" "] = None
raw_data["Professors"][raw_data["Professors"]==" "] = None
raw_data["Descriptions"][raw_data["Descriptions"]==" "] = None
```

```python
raw_data.dropna(axis=0 , inplace=True)
raw_data.head()
```

```python
raw_data.describe()
```

```python
raw_data.groupby(["Departments"]).describe()
```

```python
raw_data.head()
```

```python
text = []
for i in range(0,len(raw_data)):
    temp = "The " + raw_data.iloc[i]["Course_title"] +" course represented at the "  +  raw_data.iloc[i]["Departments"] + " by " + raw_data.iloc[i]["Professors"]+". " + raw_data.iloc[i]["Descriptions"]
    text.append(temp)
```

```python
text[0]
```

```python
import nltk
nltk.download('punkt')

from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

  
ps = PorterStemmer()


stem_text = []

for sentence in text:  
    sen = []
    words = word_tokenize(sentence)
    for word in words:
        sen.append(ps.stem(word))

    stem_text.append(sen)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')  
lemmatizer = WordNetLemmatizer()

lem_text = []

for sentence in stem_text:  
    sen = []
    for word in sentence:
        sen.append(lemmatizer.lemmatize(word))
    lem_text.append(sen)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

clean_text = []

for sentence in lem_text:  
    sen = []
    for word in sentence:
        if word not in stop_words:
            sen.append(word)
    clean_text.append(sen)
```

part 2

```python
clean_text = [' '.join(i) for i in clean_text]

whole_cleaned = str("")

for sen in clean_text:
  whole_cleaned = whole_cleaned + sen
```

```python
!pip install keybert
from keybert import KeyBERT

kw_model = KeyBERT()
keywords = kw_model.extract_keywords(whole_cleaned,top_n = 50)
```

```python
words =list(zip(*keywords))[0]
importance =list(zip(*keywords))[1]


import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(words[0:5] , importance[0:5])
```

```python
keywords
```

```python
#!pip install wordcloud
from wordcloud import WordCloud


def Convert(tup, di):
    di = dict(tup)
    return di
keywords_dictionary = {}  
keywords_dictionary = (Convert(keywords, keywords_dictionary))

wordcloud = WordCloud()
wordcloud.fit_words(keywords_dictionary)
plt.imshow(wordcloud)
```

part 3

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer( vocabulary = keywords_dictionary.keys())
X = vectorizer.fit_transform(clean_text)
```

```python
X.toarray()
```

```python
df = pd.DataFrame(X.toarray(), columns = keywords_dictionary.keys())
df = df.astype(bool)
df.head()
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.005,use_colnames=True)
```
